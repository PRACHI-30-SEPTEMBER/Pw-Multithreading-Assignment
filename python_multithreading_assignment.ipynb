## Q1. What is multithreading in Python? Why is it used? Name the module used to handle threads in Python.

**Answer:**
Multithreading is a technique where multiple threads run concurrently within a single program. Each thread performs a separate task, allowing the program to run tasks simultaneously.

**Why it is used:**
- Improves performance
- Allows concurrent execution of tasks
- Useful for I/O operations such as file handling or network operations

**Module used:** Python provides the **threading** module to handle threads.


```python
import threading
print('Threading module imported successfully')
```

## Q2. Why threading module is used? Write the use of the following functions:
1. activeCount()
2. currentThread()
3. enumerate()

**Answer:**

- **activeCount()**: Returns the number of currently active threads.
- **currentThread()**: Returns the current thread object.
- **enumerate()**: Returns a list of all active thread objects.


```python
import threading

print('Active Threads:', threading.active_count())
print('Current Thread:', threading.current_thread().name)
print('All Threads:', threading.enumerate())
```

## Q3. Explain the following functions:
1. run()
2. start()
3. join()
4. isAlive()

**Answer:**

- **run()**: Contains the code that the thread executes.
- **start()**: Starts the thread and calls the run() method internally.
- **join()**: Waits for a thread to finish before continuing execution.
- **isAlive()**: Checks whether a thread is still running.


## Q4. Write a Python program to create two threads. Thread one must print the list of squares and thread two must print the list of cubes.


```python
import threading

numbers = [1,2,3,4,5]

def print_squares():
    for n in numbers:
        print('Square:', n*n)

def print_cubes():
    for n in numbers:
        print('Cube:', n*n*n)

t1 = threading.Thread(target=print_squares)
t2 = threading.Thread(target=print_cubes)

t1.start()
t2.start()

t1.join()
t2.join()

print('Both threads finished execution')
```

## Q5. State advantages and disadvantages of multithreading.

**Advantages:**
- Improves performance of applications
- Allows multiple tasks to run simultaneously
- Efficient CPU utilization

**Disadvantages:**
- Difficult to debug
- Possibility of race conditions
- Higher complexity in program design


## Q6. Explain deadlocks and race conditions.

**Deadlock:**
A deadlock occurs when two or more threads wait for each other to release resources, causing the program to stop progressing.

**Race Condition:**
A race condition occurs when multiple threads access shared data at the same time and the final result depends on the timing of execution.
